```python
import requests

Text = "Морос и солнцэ; день чудестный!\
        Еще ты дремлешь, друг прелестный —\
        Пора, красавица, проснись:\
        Аткрой сомкнуты негой взоры\
        Навстречу севирной Авроры,\
        Звездою севира явись!\
        Вечор, ты помнешь, въюга злилась,\
        На мутном небе мгла насилась;\
        Луна, как бледное пятно,\
        Сквозь тучи мрачные жилтела,\
        И ты печальная сидела —\
        А нынче… погляди в акно:\
        Под галубыми небесами\
        Велеколепными каврами,\
        Блистя на солнце, снег лежит;\
        Прозрачный лес один чернеет,\
        И ель сквозь иней зелинеет,\
        И речка подо льдом блестит.\
        Вся комнота янтарным блеском\
        Озарена. Веселым треском\
        Трещит затопленная печь.\
        Приятно думать у лежанки.\
        Но знаешь: не велеть ли в санки\
        Кобылку бурую запречь?\
        Скользя по утреннему снегу,\
        Друг милый, предадимся бегу\
        Нетерпеливого коня\
        И навестим поля пустые,\
        Леса, недавно столь густые,\
        И берег, милый для меня."

check_text = requests.post('http://localhost:1111/model',json={'x':[Text]})
text = str(check_text.json())
print(text)
```

```python
from natasha import (
    Segmenter,
    MorphVocab,
    
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,
    
    PER,
    NamesExtractor,

    Doc
)

segmenter = Segmenter()
morph_vocab = MorphVocab()

emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)
ner_tagger = NewsNERTagger(emb)

names_extractor = NamesExtractor(morph_vocab)

doc = Doc(text)
```

```python
doc.segment(segmenter)
print(doc.tokens[:5])
print(doc.sents[:5])
```

```python
doc.tag_morph(morph_tagger)
print(doc.tokens[:5])
doc.sents[0].morph.print()
```

```python
for token in doc.tokens:
     token.lemmatize(morph_vocab)
    
print(doc.tokens[:5])
{_.text: _.lemma for _ in doc.tokens}
```
